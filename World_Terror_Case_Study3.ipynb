```python
terror = pd.read_csv('terror.csv',sep = ',',header=0,encoding = 'latin',low_memory=False)

terror.rename(columns={'eventid':'eventid','iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive','suicide':'suicide'},inplace=True)
terror=terror[['eventid','Year','Month','Day','Country','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive','suicide']]
terror['casualities']=terror['Killed']+terror['Wounded']
terror.head(10)
```

```python
tdata = terror

tdata.isnull().sum()
```
